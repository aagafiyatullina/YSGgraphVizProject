```python
def findparents(child):
    interactions = content1.loc[content1['user'] == child]
    all_parents = []
    for index, row in interactions.iterrows():
        parent_interactions = comments_forest1.loc[comments_forest1['comm_id']==row['c_id']]
        if (parent_interactions.empty==False):
            parent_comment = content1.loc[content1['c_id']==parent_interactions['parent_id'].iloc[0]]
            if (parent_comment.empty==False):
                parent_name = parent_comment['user'].iloc[0]
                all_parents.append(parent_name)
    return all_parents
```

средняя глубина комментария

```python
def comm_depth(comment):
    depth = 0
    prev_comment = comments_forest1.loc[comments_forest1['comm_id']==comment]
    if(prev_comment.empty==False):
        parent_comment = prev_comment['parent_id'].iloc[0]
    while(prev_comment.empty == False):
        depth += 1
        prev_comment = comments_forest1.loc[comments_forest1['comm_id']==parent_comment]
        if (prev_comment.empty==False):
            parent_comment = prev_comment['parent_id'].iloc[0]
    return depth
```

```python
avg_depth = []
```

```python
for index, row in users1.iterrows():
    curr_user = row['user']
    all_comm = content1.loc[(content1['user']==curr_user) & (content1['type']==1)]
    amount = len(all_comm.index)
    sum_of = 0
    for index1, row1 in all_comm.iterrows():
        sum_of += comm_depth(row1['c_id'])
    avg = 0.0
    if (amount != 0):
        avg = sum_of / amount
    print(avg)
    avg_depth.append(avg)
```

6) процент комментариев без ответа

```python
no_answer = []
```

```python
def unanswered(comment):
    children = comments_forest.loc[comments_forest['parent_id']==comment]
    if (len(children.index)==0):
        return True
    else:
        return False
```

```python
king = 'kxx'
allc = content1.loc[(content1['user']==king) & (content1['type']==1)]
amount = len(allc.index)
noans = 0
for index, row in allc.iterrows():
    if (unanswered(row['c_id']) == True):
        noans += 1
avg = 0.0
avg = noans / amount
print(avg)
```
